```python
import csv
import matplotlib.pyplot as plt
import numpy as np
from scipy.signal import argrelextrema

with open('半外腔縱模.csv', encoding='utf-8') as f:
    reader = csv.reader(f, delimiter=',')
    header = next(reader)
    times = []
    ch1s = []
    ch2s = []
    for row in reader:
        time, ch1, ch2 = row
        times.append(float(time))
        ch1s.append(float(ch1))
        ch2s.append(float(ch2))

times = np.array(times)
ch1s = np.array(ch1s)
ch2s = np.array(ch2s)
```

```python
plt.plot(times, ch2s)
plt.plot(times, ch1s*1e-2)
argrelextrema(ch2s, np.greater)
```
